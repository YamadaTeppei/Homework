### 1. P = NP ですか? 

現在の段階ではP≠NPです。
何故ならば現段階ではPとは多項式時間で解ける問題のことでありNPは指数時間でしか解くことができない問題であるからです。またNP問題の中でも特に難しい問題をNP完全問題と言います。そしてPとNPの関係は今だにP≠NP、P=NPのどちらも証明がされていません。

### 4. 何故Googleは, Go言語を開発したのだと思いますか?

大企業が独自のプログラミング言語を使うのは、商業的に見てそちらの方が自社のサービス運営で利益になるからだと思います。
何故なら独自の言語はその会社にとって、もっとも効率良くサービス運営できるようになっていると思われます。また会社内でその言語について分からないことがなく、社員の教育に関しても楽になり、オープンソースにする事により社外からの意見を聞くことがでいき市場の流れを変える時の足がかりにもなり得ると思います。

### 5.  何故自動運転車や配達ドローンの実用化が滞っているのだと思いますか?

自動運転も配達ドローンも実用化しようと思えば可能な状態まで開発は進んでいると思われますがどちらもコストや安全性やもしもの時の責任の在りかなどにより滞っているのだと思います。

まずコストに関してはどちらも制作費が現状では高く量産に向かない状態であると思います。また今まで運転や配達を仕事にしていた人達の仕事が減ることとなるので、その辺りに対する対策も倫理的にしなければならないと思います。

次に安全性ですがこちらは今はまだ発展途上であり市場に出して良い状態になっていないと思います。例えば自動運転であれば一つの認識ミスで大事故になりますし、配達ドローンであれば風の強弱や荷物の固定の仕方でこちらも人死が出る可能性があります。またどちらも通信技術が用いられ電気的な制御によるものなので、遠隔操作や細工をする事によるテロなどの犯罪対策もまだ人が運転するより安全であると言えない状態だと思いんます。

最後に責任の在りかですがこちらは基本的には企業が責任を追う事になると思います。

こういった事により企業としてはまだ製品化すると損失の方が大きくなる可能性が高すぎる事により滞っているのだと思います。
