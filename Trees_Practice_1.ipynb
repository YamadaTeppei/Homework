## Problem 0  - Create Node/BinaryTree Class

木構造の各要素を表す **Node** クラスと, 二分木を表す **BinaryTree** クラスを実装してみましょう.  

### Node Class

Nodeクラスは以下の変数を持ち, メソッドは持ちません.  インスタンス生成時には, valには与えられた値を, leftとrightには *None* をセットします.  

    1. val: ノードの持つデータ
    2. left: 左の子ノードへのポインター
    3. right: 右の子ノードへのポインター

```python
class Node:
    def __init__(self,val):
        self.right = None
        self.left = None
        self.val = val
```

### BinaryTree Class

BinaryTreeクラスは以下の変数を持ちます.  

    1. root: 根ノードへのポインター

```python
class BinaryTree:
    def __init__(self,val):
        self.root = Node(val)
        
    def countNode(self,root):
        if root is None:
            return 0
        return 1 + self.countNode(root.left) + self.countNode(root.right)
        
    def getHeight(self,root):
        if root is None:
            return -1
        left_height = self.getHeight(root.left)
        right_height = self.getHeight(root.right)
        
        return 1 + max(left_height, right_height)
        
    def existsAt(self,data):
        queue = []
        queue.append(self.root)
        while len(queue) > 0:
            currNode = queue.pop(0)
            if currNode.val == data:
                return True
            if currNode.left != None:
                queue.append(currNode.left)
            if currNode.right != None:
                queue.append(currNode.right)
        return False
            # レベルの算出がまだできていません。
            
        
            
        
#     def convertToList(self,root,i = 0,tree_list = []):
#         if root:
#             x = []
#             x.append(root.val)
#             if  i == 0:
#                 tree_list.append(x)
#             if i == 1:
#                 tree_list[0].insert(0,x)
#             if i == 2:
#                 tree_list[len(tree_list)-1].append(x)
#             self.convertToList(root.left,1,tree_list)
#             self.convertToList(root.right,2,tree_list)
#             return tree_list

    def printNode(self,root):
        if root:
            print(root.val, end = " ")
            self.printNode(root.left)
            self.printNode(root.right)
```

## Problem 3 - Binary Heap

**Binary Heap (二分ヒープ)** とは,  親ノードと子ノードの値の大小によって定義される完全二分木のことです.  
二分ヒープには, 根ノードが最小値である **Min Heap(最小ヒープ)** と, 根ノードが最大値である **Max Heap(最大ヒープ)** の2種類があります.  
実装には, 空間計算量の観点から **配列(リスト)** を用いた (i.e. Nodeを用いない) 実装がされることが多いです.  

ここでは, 最小二分ヒープを表す **MinBinaryHeapクラス** と, 以下の基本操作を実装してみましょう.  ただし, コンストラクターは以下のセルの通りとします.


### #1: **size()**  

    二分ヒープ内のノード数を返す関数

### #2: **get(i)**

    配列におけるi番目のノードを返す関数
    
### #3: **parent(i)**

    配列におけるi番目のノードの親ノードのインデックスを返す関数

### #4:  **left(i)**

    配列におけるi番目のノードの左の子ノードのインデックスを返す関数 

### #5: **right(i)**

    配列におけるi番目のノードの右の子ノードのインデックスを返す関数 
    
### #6: **getMin()**

    二分ヒープ内の最小値(つまり根ノードの値)を返す関数
    
### #7: **minHeapify(i)**

    配列におけるi番目のノード以降の並びを, 最小ヒープの性質(親ノード<=子ノード)に基づいて並び替える関数 
    
### #8: **insert(a)**

    与えられた値 a を二分ヒープ内の適切な場所に挿入する関数 
    
### #9: **deleteMin()**

    二分ヒープから最小値(根ノード)を削除し, 必要に応じてノードを並び替える関数 
    
### #10: **delete(a)**

    与えられた値 a を二分ヒープから削除し, 必要に応じてノードを並び替える関数

```python
class MinBinaryHeap:
    # Constructor
    def __init__(self):
        self.nodes = []
        
    def size(self):
        return len(self.nodes)
    
    def get(self,i):
        if i <= self.size():
            return self.nodes[i - 1]
        else:
            return "erorr: index(i) no find"
        
    def parent(self,i):
        if i == 1:
            return "erorr: i is root_node" 
        elif i <= self.size():
            return self.nodes[(i  // 2) - 1]
        else:
            return "erorr: index(i) no find"
        
    def left(self,i):
        if i*2 > self.size():
            return "eroor: i is leaf_node"
        elif i <= self.size():
            return self.nodes[(2 * i) - 1]
        else:
            return "erorr: index(i) no find"
        
    def right(self,i):
        if i*2 > self.size():
            return "erorr: i is leaf_node"
        elif i <= self.size():
            return self.nodes[(2 * i + 1) - 1]
        else:
            return "erorr: index(i) no find"
        
    def getMin(self):
        return self.nodes[0]
        
    def minHeapify(self,i):
        minHeapify_nodes = self.nodes[i-1:]
        def inner(nodes,i = 1):
            # i番目以降の配列内で親ノードになる配列分繰り返す
            if i*2 < len(nodes):
                #子ノードの左と右を比較し右の方が小さければ入れ替える
                if nodes[(i * 2)-1] > nodes[(i * 2 + 1)-1]:
                    x = nodes[(i+1)-1]
                    nodes[(i+1)-1] = nodes[(i+2)-1]
                    nodes[(i+2)-1] = x
                #親ノードと左の子ノードを比較し子ノードの方が小さければ入れ替える
                if nodes[(i)-1] > nodes[(i * 2)-1]:
                    x = nodes[(i)-1]
                    nodes[(i)-1] = nodes[(i*2)-1]
                    nodes[(i*2)-1] = x
                    
                inner(nodes,i + 1)
                return nodes
        return inner(minHeapify_nodes)
            
        
    def insert(self,data):
        self.nodes.append(data)
        x = self.size()
        #入れたデータと親ノードを比較しデータの方が小さければ入れ替える
        while x <= 0:
            if self.nodes[x] < self.nodes[x//2]:
                y = self.nodes[x//2]
                self.nodes[x//2] = self.nodes[x]
                self.nodes[x] = y
            x //= 2
        
    def deleteMin(self):
        self.nodes = self.minHeapify(2)
        
    def delete(self,data):
        for i in range(self.size()-1):
            if self.nodes[i] == data:
                x = self.minHeapify((i+1)+1)
                self.nodes = self.nodes[:(i-1)+1]
                self.nodes.extend(x)
                break
        else:
            return"erorr: "
        
```
